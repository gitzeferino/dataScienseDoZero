```python
# Python: Tipos de Dados

# 1. Inteiros (int)
# Os inteiros são números sem parte decimal.
inteiro1 = 10
inteiro2 = -25

# Exemplo de operações com inteiros
soma = inteiro1 + inteiro2
multiplicacao = inteiro1 * inteiro2

print("1. Inteiros (int):")
print(f"Inteiro1: {inteiro1}, Inteiro2: {inteiro2}")
print(f"Soma: {soma}, Multiplicação: {multiplicacao}\n")


# 2. Ponto Flutuante (float)
# O tipo float é usado para números com parte decimal.
flutuante1 = 3.14
flutuante2 = -7.5

# Exemplo de operações com floats
divisao = flutuante1 / flutuante2
subtracao = flutuante1 - flutuante2

print("2. Ponto Flutuante (float):")
print(f"Flutuante1: {flutuante1}, Flutuante2: {flutuante2}")
print(f"Divisão: {divisao}, Subtração: {subtracao}\n")


# 3. Strings (str)
# Strings são sequências de caracteres.
nome = "João"
mensagem = "Olá, " + nome + "!"

# Acessando elementos de uma string
primeira_letra = nome[0]

print("3. Strings (str):")
print(f"Nome: {nome}")
print(f"Mensagem: {mensagem}")
print(f"Primeira letra do nome: {primeira_letra}\n")


# 4. Listas (list)
# Listas são coleções ordenadas e mutáveis.
lista = [1, 2, 3, 4, 5]

# Adicionando um novo elemento à lista
lista.append(6)

# Acessando elementos da lista
terceiro_elemento = lista[2]

print("4. Listas (list):")
print(f"Lista: {lista}")
print(f"Terceiro elemento da lista: {terceiro_elemento}\n")


# 5. Tuplas (tuple)
# Tuplas são coleções ordenadas, mas imutáveis.
tupla = ("João", 30, "Programador")

# Acessando elementos da tupla
nome_tupla = tupla[0]

print("5. Tuplas (tuple):")
print(f"Tupla: {tupla}")
print(f"Nome da tupla: {nome_tupla}\n")


# 6. Dicionários (dict)
# Dicionários armazenam pares chave-valor.
dicionario = {
    "nome": "João",
    "idade": 30,
    "curso": "Programação"
}

# Acessando valor a partir da chave
nome_dict = dicionario["nome"]

print("6. Dicionários (dict):")
print(f"Dicionário: {dicionario}")
print(f"Nome do dicionário: {nome_dict}\n")


# 7. Conjuntos (set)
# Conjuntos são coleções de elementos únicos e não ordenados.
conjunto = {"maçã", "banana", "laranja"}

# Adicionando um novo elemento ao conjunto
conjunto.add("uva")

# Tentando adicionar um elemento duplicado
conjunto.add("banana")

print("7. Conjuntos (set):")
print(f"Conjunto: {conjunto}\n")


# 8. Booleanos (bool)
# Booleanos representam valores lógicos: True ou False.
is_active = True
is_completed = False

# Exemplo de comparação
is_equal = (inteiro1 == inteiro2)

print("8. Booleanos (bool):")
print(f"Ativo: {is_active}, Completo: {is_completed}")
print(f"Comparação: {is_equal}\n")


# 9. Tipos de Dados Complexos (complex)
# Os números complexos têm uma parte real e uma parte imaginária.
numero_complexo = 3 + 4j

# Acessando a parte real e imaginária
parte_real = numero_complexo.real
parte_imaginaria = numero_complexo.imag

print("9. Tipos de Dados Complexos (complex):")
print(f"Número complexo: {numero_complexo}")
print(f"Parte real: {parte_real}, Parte imaginária: {parte_imaginaria}\n")


# 10. Tipo None
# O tipo None é utilizado para representar a ausência de valor.
variavel_nula = None

print("10. Tipo None:")
print(f"Valor de variavel_nula: {variavel_nula}\n")

# Finalizando o notebook com uma mensagem
print("Fim do notebook sobre tipos de dados em Python.")
```
